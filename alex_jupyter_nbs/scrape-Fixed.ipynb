```python
movies = list(all_movies.Movie.values)
rem = [x for x in brands['replace'].values if x not in movies]
box_office = []


for x in rem:
    
    text = str(x) + ' box office'
    url = 'https://google.com/search?q=' + text
    page = requests.get(url)
    soup = BeautifulSoup(page.content)

    
    
    #driver.get('https://www.google.com/')
    #search = driver.find_element_by_name('q')
    #search.send_keys(x + ' box office')
    #search.send_keys(Keys.RETURN)
    
    #page = requests.get(driver.current_url)
    #soup = BeautifulSoup(page.content)
    
    try:
        res = str(soup)[str(soup).index('USD')-20:].split('>')[1]
        if 'USD' in res:
            box_office.append((x, res))
        else:
            continue
    except:
        continue
```
