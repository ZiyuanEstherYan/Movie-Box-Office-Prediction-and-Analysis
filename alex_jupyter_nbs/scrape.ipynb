```python
movies = list(all_movies.Movie.values)
rem = [x for x in brands['replace'].values if x not in movies]
box_office = []

for x in rem:
    driver.get('https://www.google.com/')
    search = driver.find_element_by_name('q')
    search.send_keys(x + ' box office')
    search.send_keys(Keys.RETURN)
    
    page = requests.get(driver.current_url)
    soup = BeautifulSoup(page.content)
    
    try:
        res = str(soup)[str(soup).index('USD')-20:].split('>')[1]
        if 'USD' in res:
            box_office.append((x, res))
        else:
            continue
    except:
        continue
```

```python
box_office_fix = [int(re.sub('(\\xa0million USD)|(USD)|(\.)|(\,)|(<\/div)|(<\/span)', '', x))*1000000 if 'million' in x else int(re.sub('(\\xa0billion USD)|(USD)|(\.)|(\,)|(<\/div)|(<\/span)', '', x))*1000000000 if 'billion' in x else int(re.sub('(\,)|(<\/div)|(USD)', '', x))for x in [x[1] for x in box_office]]
```
