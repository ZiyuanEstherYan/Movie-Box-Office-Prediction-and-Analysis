```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import pairwise_distances
```

```python
dat = pd.read_csv('data/movies.csv')
dat = dat.drop('X1', axis = 1)
filters = ['replace', 'brand_ods_id', 'year_bin', 'franchise', 'budget_quant', 'rating', 'source']
df = dat[filters]
df = pd.get_dummies(df, columns = ['year_bin', 'franchise', 'budget_quant', 'rating', 'source'])
df = pd.concat([df, dat.iloc[:, 21:] ], axis = 1)
df = df.dropna()
```

```python
df2 = df.T
df2.columns = df2.iloc[0]
df2 = df2.drop('replace')
```

```python
jac_sim = 1 - pairwise_distances(df2.T, metric = "hamming")
jac_sim = pd.DataFrame(jac_sim, index=df2.columns, columns=df2.columns)
```
